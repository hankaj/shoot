```python
# Import libraries
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
# Get competitions data
# Dataframe: SIUS results, sport shooting, air pistol, Klaudia Breś, 2015-2021
df = pd.read_csv('export_dataframe.csv', index_col='date', parse_dates=True)
# Show the data
df
```

```python
# Get the number of rows in columns in the data set
df.shape
```

```python
# Plot results against number of inner tens
plt.figure(figsize=(16,8))
plt.title('Results history')
plt.plot(df['inner_tens'], df['n_points'], 'ro')
plt.xlabel('Number of inner tens', fontsize=18)
plt.ylabel('Result', fontsize=18)
plt.show()
```

```python
# Visualize results history
plt.figure(figsize=(16,8))
plt.title('Results history')
plt.plot(df['n_points'])
plt.xlabel('Date', fontsize=18)
plt.ylabel('Points', fontsize=18)
plt.show()
```

```python
mean_result = pd.DataFrame(df.mean()[0:60])
mean_result.columns = ['result']
mean_result
```

```python
# Plot average shooting
plt.figure(figsize=(24,8))
plt.title('Average shooting')
plt.plot(mean_result['result'])
plt.xlabel('Shot index', fontsize=18)
plt.ylabel('Result', fontsize=18)
plt.show()
```
